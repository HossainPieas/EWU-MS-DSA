```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

***Importing & Understanding The Dataset***

```python
#importing the dataset which is already in the folder
df = pd.read_csv('machine_sensor_dataset.csv') 
```

```python
#find out what the columns are in this dataset
df.columns
```

```python
#reading first few (10) lines of the dataset
df.head(10)
```

```python
#reading the last line of the dataset to understand if there is anything different
df.tail(2)
```

```python
#Find out crucial info about the dataset
df.info()
```

We find that there are 60 columns which are float values and only 1 column, which is class, has integer values. All columns mentioning features do have missing values. Class does not have any missing values and is the only complete column in this dataframe.

```python
df.isnull().sum()
```

From here we understood how many null values there are in each column.

```python
#as there are a lot of missing values, dropping such rows would lead to less data, instread we can fill the missing values with mean
#fill missing values with mean as it will not skew the data
df = df.fillna(df.mean(), inplace=False)
print(df)
```

```python
#sort the data by class to make the data more readable
df=df.sort_values(by=['Class'],ascending=True)
print(df)
```

```python
#Find out summary statistics of the dataset
df.describe()
```

```python
df_grouped = df.groupby('Class').mean()
print(df_grouped)
```

```python
df.corr()
```

```python
# Class distribution
class_counts = df['Class'].value_counts()
print(class_counts)

# Plot class distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df, palette='Set2')
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution')
plt.show()
```
