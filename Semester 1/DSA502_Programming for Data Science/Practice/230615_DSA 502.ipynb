## All the numeric operations (One way)

```python
num_1=input("Input the first number \t")
num_2=input("Input the second number \t")
num_1=int(num_1)
num_2=int(num_2)
sum= num_1 + num_2
mul= num_1 * num_2
div= num_1 / num_2
sub= num_1 - num_2
mod= num_1 % num_2
div_res= num_1 // num_2
sqr= num_1 ** num_2

print(f"The sum of {num_1} and {num_2} is {sum}. \nThe multiplication of {num_1} and {num_2} is {mul}. \nThe div of {num_1} and {num_2} is {div}. \nThe difference of {num_1} and {num_2} is {sub}. \nThe mod of {num_1} and {num_2} is {mod}. \nThe remainder of {num_1} and {num_2} is {div_res}. \nThe power of {num_1} and {num_2} is {sqr}. \n")
```

## All the numeric options (Other way)

```python
num_1=float(input("Input the first number \t"))
num_2=float(input("Input the second number \t"))
sum= num_1 + num_2
mul= num_1 * num_2
div= num_1 / num_2
sub= num_1 - num_2
mod= num_1 % num_2
div_res= num_1 // num_2
sqr= num_1 ** num_2

print(f"The sum of {num_1} and {num_2} is {round(sum,2)}. \nThe multiplication of {num_1} and {num_2} is {round(mul,2)}. \nThe div of {num_1} and {num_2} is {round(div,2)}. \nThe difference of {num_1} and {num_2} is {round(sub,2)}. \nThe mod of {num_1} and {num_2} is {round(mod,2)}. \nThe remainder of {num_1} and {num_2} is {round(div_res,2)}. \nThe power of {num_1} and {num_2} is {round(sqr,2)}. \n")
```

## If-elif-else

```python
num1=input("Input first number \n")
num2=input("Input second number \t")
num1=float(num1)
num2=float(num2)
if num1>num2:
  print (f"{num1} is the largest")
elif num1==num2:
  print(f"{num1} is equal to {num2}")
else:
  print(f"{num2} is the largest")
```

```python
num1=input("Divisible by the number \n")
num2=input("Number to be checked \t")
num1=int(num1)
num2=int(num2)

if num1<=num2 and num2%num1==0:
  print(f"{num2} is divisible by {num1}")
elif num1>num2:
  print(f"{num2} is smaller than {num1}")
else:
  print(f"{num2} is not divisible by {num1}")
```

## While Loop

## For Loop

```python
gpa=0.0

num_of_students=int(input("Enter the no of students in your class \t"))
for i in range(0,num_of_students): #For loop for students
  num_of_course=int(input("Number of courses \n"))

  for i in range(0,num_of_course): #0 to n-1 #For loop of numbers
    marks=float(input("Enter your marks \t"))

    if marks<=100 and marks >90:
      gpa=gpa+4.0
    elif marks>=80:
      gpa=gpa+3.75
    else:
      gpa=gpa+0.00
  cgpa=gpa/num_of_course

  print(f"Your total GPA is {gpa}. Your total CGPA is {cgpa}")
```
